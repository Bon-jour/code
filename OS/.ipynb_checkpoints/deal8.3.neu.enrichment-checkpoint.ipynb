```r
library(Seurat)
```

```r
sce = readRDS('neu.final.rds')
```

```r
DimPlot(sce,reduction = 'umap',label = T)
```

```r
Idents(sce) = 'seurat_clusters'
```

```r
library(CytoTRACE)
```

```r
#  https://www.jianshu.com/p/0f1e5479e72f   cytotrace 完全颠倒的原因
```

```r
exp1 <- as.matrix(sce@assays$RNA@counts)
exp1 <- exp1[apply(exp1 > 0,1,sum) >= 5,]
```

```r
results <- CytoTRACE(exp1,ncores = 10)
```

```r
phenot <- sce$seurat_clusters
phenot <- as.character(phenot)
names(phenot) <- rownames(sce@meta.data)
emb <- sce@reductions[["umap"]]@cell.embeddings
```

```r
options(repr.plot.width = 12)
plotCytoTRACE(results, phenotype = phenot, emb = emb, outputDir = './')
plotCytoGenes(results, numOfGenes = 30, outputDir = './')
```
