```r
merged = as.SingleCellExperiment(sce)

abundances <- table(merged$celltype, merged$sample) 
abundances <- unclass(abundances) 

library(scater)
library(SingleCellExperiment)

colLabels(merged) <- sce$celltype
```

```r
merged
```

```r
library(DESeq2)
library(limma)
library(edgeR)
```

```r

table(colLabels(merged), merged$group)

extra.info <- colData(merged)[match(colnames(abundances), merged$sample),]

y.ab <- DGEList(abundances, samples=extra.info)

keep <- filterByExpr(y.ab, group=y.ab$samples$group)
y.ab <- y.ab[keep,]

design <- model.matrix(~factor(celltype) + factor(group), y.ab$samples)
y.ab <- estimateDisp(y.ab, design, trend="none")
```

```r


fit.ab <- glmQLFit(y.ab, design, robust=TRUE, abundance.trend=FALSE)
res <- glmQLFTest(fit.ab, coef=ncol(design))
summary(decideTests(res))

    diff=topTags(res)$table

diff

padj = 0.05
foldChange = 0


    diff$group = as.factor(ifelse(diff$FDR < padj & abs(diff$logFC) >= foldChange, 
                                 ifelse(diff$logFC> foldChange ,'Up','Down'),'NoSignifi'))
    table(diff$group)
    res = diff

library("edgeR")  
    library(dplyr)
    library(ggrepel)
    library(ggplot2)
    library(cowplot)
```

```r
setwd('/import/GSE213982_fel/')
library(DIALOGUE)
library(Seurat)
sce = readRDS('./female.rds')
table(sce$celltype)
```

```r
Idents(sce) <- "celltype"
# Downsample the number of cells per identity class
#ob1 <- subset(x = sce, downsample = 000)
ob1 = sce
library(Seurat)
ob1 <- FindVariableFeatures(ob1, selection.method = "vst", nfeatures = 2000)
ob1
```

```r
set.seed(1263)
# Simulating data based on Seurat object of PBMCs
obj<-ob1
obj@meta.data$samples <- obj@meta.data$sample
obj@meta.data$cell.subtypes<-obj@meta.data$celltype
b1<-is.element(obj@meta.data$cell.subtypes,c("Ast",'InN'))
b2<-is.element(obj@meta.data$cell.subtypes,c("Mic","Mix"))
b3<-is.element(obj@meta.data$cell.subtypes,c('End', 'ExN'))
b4<-is.element(obj@meta.data$cell.subtypes,c("OPC","Oli"))




obj@meta.data$samples[b1] <- sample(obj@meta.data$sample, size = sum(b1), replace =FALSE)
obj@meta.data$samples[b2] <- sample(obj@meta.data$sample, size = sum(b2), replace =FALSE)
obj@meta.data$samples[b3] <- sample(obj@meta.data$sample, size = sum(b3), replace =FALSE)
obj@meta.data$samples[b4] <- sample(obj@meta.data$sample, size = sum(b4), replace =FALSE)

DIALOGUE_make.cell.type.seurat<- function(obj, cell.subtypes, name){
  # Given a Seurat object with "cell.subtypes" and "samples" information provided in the meta.data field
  
  if(is.null(obj@meta.data$cell.subtypes)){return("Cell subtypes information is missing")}
  if(is.null(obj@meta.data$samples)){return("Sample information is missing")}
  
  sub_obj<- subset(x = obj,cells = rownames(obj@meta.data)[is.element(obj@meta.data$cell.subtypes,cell.subtypes)])
  sub_obj<- NormalizeData(sub_obj,normalization.method = "LogNormalize", scale.factor = 100000)
  sub_obj<-FindVariableFeatures(sub_obj,selection.method = "vst", nfeatures = 2000)
  sub_obj<-ScaleData(sub_obj)
  sub_obj<-RunPCA(sub_obj,npcs = 30) 
  
  tpm<- as.matrix(sub_obj@assays$RNA@data)
  
  X<- sub_obj@reductions$pca@cell.embeddings
  assertthat::are_equal(rownames(X),colnames(tpm))
  n_cells<- ncol(tpm)
  samples<- sub_obj@meta.data$sample
  n_count<- as.vector(scale(log(sub_obj@meta.data$nCount_RNA)))
  metadata<- data.frame(nFeatures = sub_obj@meta.data$nFeature_RNA,
                        samples = samples,
                        group = sub_obj@meta.data$group,
                        cell.subtypes = as.character(sub_obj@meta.data$cell.subtypes))
  r<-make.cell.type(name = name ,tpm,samples,X,metadata, cellQ = n_count)
  return(r)
}



```

```r
r1<- DIALOGUE_make.cell.type.seurat(obj, cell.subtypes = c('ExN'), name = "A")
r2<- DIALOGUE_make.cell.type.seurat(obj, cell.subtypes = c("Ast"), name = "B")
rA<- list(A = r1,B = r2)
library(DIALOGUE)
results.dir = './'
```

```r
R <- DIALOGUE.run(rA = rA, main = 'res',# list of cell.type objects
                  param = DLG.get.param( pheno = "group", # number of MCPs to identify
                                        results.dir = results.dir,
                                        spatial.flag = F,plot.flag = T,
                                        conf = "cellQ"))
```

```r
par(mfrow=c(1,2),oma = c(8, 1, 0, 5),xpd = T)
summary(R)

print(round(R$phenoZ,2))
R$MCPs
```
