```r
library(miloR)
library(SingleCellExperiment)
library(scater)
library(scran)
library(dplyr)
library(patchwork)
```

```r
sce <- as.SingleCellExperiment(sce)
 
#Visualize the data
plotReducedDim(sce, colour_by="celltype", dimred = "TSNE") 
```

```r
scmilo <- Milo(sce)
 
#Construct KNN graph
scmilo <- buildGraph(scmilo, k = 10, d = 10,reduced.dim = "PCA")
```

```r

 
#Defining representative neighbourhoods on the KNN graph
set.seed(10)
scmilo <- makeNhoods(scmilo, prop = 0.8,
                     k = 30, d = 30, 
                     refined = TRUE, reduced_dims = "PCA")
plotNhoodSizeHist(scmilo)
 
#Counting cells in neighbourhoods
scmilo <- countCells(scmilo, 
                     meta.data = as.data.frame(colData(scmilo)), 
                     sample = "sample")
```

```r
sc_design <- data.frame(colData(scmilo))[,c("sample", "Condition")]
 
sc_design <- distinct(sc_design)
rownames(sc_design) <- sc_design$sample
```

```r
scmilo <- calcNhoodDistance(scmilo, d = 10, reduced.dim = "PCA")
```

```r
results <- testNhoods(scmilo, design = ~ Condition, design.df = sc_design, norm.method='RLE')
```

```r
#Testing
results <- testNhoods(scmilo, design = ~ Condition, 
                      design.df = sc_design, reduced.dim="PCA")
head(results)
```

```r

 

 
results %>%
  arrange(SpatialFDR) %>%
  head() 
```

```r
scmilo <- buildNhoodGraph(scmilo)
```

```r
umap_pl <- plotReducedDim(scmilo, dimred = "UMAP", 
                          colour_by="celltype", text_by = "celltype", 
                          text_size = 3, point_size=0.5) +
  guides(fill="none")
```

```r
nh_graph_pl <- plotNhoodGraphDA(scmilo, results, layout="UMAP",alpha = 1) 
 
umap_pl + nh_graph_pl +
  plot_layout(guides="collect")
```
