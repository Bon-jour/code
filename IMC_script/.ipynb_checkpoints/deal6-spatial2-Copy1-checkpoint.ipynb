```r
label_files <- list.files("./data2", 
                          full.names = TRUE, pattern = ".rds$")
# Read in SPE objects
```

```r
spes <- lapply(label_files, readRDS)
```

```r
c = list()
```

```r
for ( i in seq(1:4)){
    
    c[[i]] = colData(spes[[i]])
    # Merge SPE objects
}
```

```r
concat_spe <- do.call("rbind", c)
```

```r
concat_spe$barcode = rownames(concat_spe)
```

```r
bre = readRDS('after.spe.rds')
```

```r
bre@colData$barcode = rownames(colData(bre))
```

```r
bre1  = bre[,bre$barcode %in% rownames(concat_spe) ]
```

```r
concat_spe = concat_spe[match(bre1@colData$barcode,concat_spe$barcode) ,]
```

```r
bre1@colData$celltype = concat_spe$celltype
```

### 空间邻域分析

```r
options(repr.plot.width = 12,repr.plot.height = 12)
```

```r
# By celltypes
spe <- aggregateNeighbors(spe, 
                          colPairName = "knn_interaction_graph", 
                          aggregate_by = "metadata", 
                          count_by = "celltype")

set.seed(220705)

cn_1 <- kmeans(spe$aggregatedNeighbors, centers = 6)
spe$cn_celltypes <- as.factor(cn_1$cluster)

plotSpatial(spe, 
            node_color_by = "cn_celltypes", 
            img_id = "patient_id", 
            node_size_fix = 0.5) +
    scale_color_brewer(palette = "Set3")
```

```r
count = table(spe$sample_id,spe$celltype) %>%  as.data.frame()
```

```r
count2 = count %>% separate(Var1, c('patient_id', 'loc'))
```

```r
unique(count2$Var2)
```

```r
library(ggpubr)
```

```r
count2$Freq = log2(count2$Freq +1)
```

```r
for (i in unique(count2$Var2)){

    new = count2 %>%  dplyr::filter(Var2 == i)
    library (ggplot2)

    #create violin plot to visualize distribution of points by team
    ggplot(new, aes(x=patient_id, y=Freq, fill=patient_id)) + 
      geom_violin() + stat_compare_means(aes(label = after_stat(p.signif)),
                     method = "t.test", ref.group = "T005") 
    
}
```

```r
i= 'TCL_CD8A_GRZB'

new = count2 %>%  dplyr::filter(Var2 == i)
library (ggplot2)

#create violin plot to visualize distribution of points by team
ggplot(new, aes(x=patient_id, y=Freq, fill=patient_id)) + 
  geom_violin() + stat_compare_means(aes(label = after_stat(p.signif)),
                 method = "t.test", ref.group = "T005") +theme_classic()

ggsave('TCL.violin.pdf',height = 5,width = 5)
```
