```r
ifnb.list <- SplitObject(s2, split.by = "orig.ident")
testAB.anchors <- FindIntegrationAnchors(object.list = ifnb.list, dims = 1:20)
testAB.integrated <- IntegrateData(anchorset = testAB.anchors, dims = 1:20)
DefaultAssay(testAB.integrated) <- "integrated"
# Run the standard workflow for visualization and clustering
testAB.integrated <- ScaleData(testAB.integrated, features = rownames(testAB.integrated))
testAB.integrated <- RunPCA(testAB.integrated, npcs = 20, verbose = FALSE)
testAB.integrated <- FindNeighbors(testAB.integrated, dims = 1:20)
testAB.integrated <- FindClusters(testAB.integrated, resolution = 0.3)
```

```r
testAB.integrated <- RunUMAP(testAB.integrated, dims = 1:20)
testAB.integrated <- RunTSNE(testAB.integrated, dims = 1:20)
scrna = testAB.integrated
```

```r
DefaultAssay(scrna)='RNA'

Idents(scrna) = 'seurat_clusters'

DotPlot(scrna,features = c('Il10','Arg1','Arg2','Lyve1','Mrc1','Cd163','Mertk'))
```

```r
DefaultAssay(outdaa) = 'integrated'
DefaultAssay(quary) = 'integrated'
```

```r
outdaa <- ScaleData(outdaa, features = rownames(outdaa))
outdaa <- RunPCA(outdaa, npcs = 20, verbose = FALSE)
quary <- ScaleData(quary, features = rownames(quary))
quary <- RunPCA(quary, npcs = 20, verbose = FALSE)
```

```r
DefaultAssay(outdaa) = 'RNA'
DotPlot(outdaa,features = c('Lyve1','Mrc1','Cd163','Mertk'))
```

```r
DefaultAssay(outdaa) = 'integrated'
DefaultAssay(quary) = 'integrated'
```

```r
anchors <- FindTransferAnchors(
  reference = outdaa,
  query = quary,
  normalization.method = "LogNormalize",
  reference.reduction = "pca",
  dims = 1:20
)
```

```r
quary <- RunUMAP(quary, dims = 1:20,return.model=TRUE)
outdaa <- RunUMAP(outdaa, dims = 1:20,return.model=TRUE)
```

```r
pbmc3k <- MapQuery(
  anchorset = anchors,
  query = quary,
  reference = outdaa,
  refdata = list(
    celltype.l1 = "celltype",
    celltype.l2 = "seurat_clusters",
    predicted_ADT = "ADT"
  ),
  reference.reduction = "pca", 
  reduction.model = "umap"
)
```

```r
DimPlot(pbmc3k, reduction = "ref.umap", group.by = "predicted.celltype.l1", label = TRUE, label.size = 3, repel = TRUE) + NoLegend()
```

```r
Idents(pbmc3k) = 'seurat_clusters'
```

```r
DimPlot(pbmc3k, reduction = "ref.umap",label = T)
```

```r
DimPlot(outdaa,label = T)
```

```r
DefaultAssay(pbmc3k) = 'RNA'


sce.markers <- FindAllMarkers(object = pbmc3k, only.pos = TRUE, 
                              min.pct = 0.25, 
                              thresh.use = 0.25)
```

```r
library(dplyr)
top5 <- sce.markers %>% group_by(cluster) %>% top_n(8, avg_log2FC)
library(ggplot2)
```

```r
DotPlot(pbmc3k, features = unique(top5$gene),
 assay='RNA' ) + coord_flip()+ theme(axis.text.x = element_text(angle = 45, 
 vjust = 0.5, hjust=0.5))
```
