```r
sce = readRDS('merge.8.9.to.0.rds')
```

```r
library(Seurat)
```

```r
Idents(sce) = 'orig.ident'
```

```r
wt3d = subset(sce,idents = 'WT3d')
```

```r
DefaultAssay(wt3d) = 'RNA'
```

```r
library(dplyr)
```

```r
expr <- wt3d@assays$RNA@data
gene_name <- c("Trem2")
```

```r
gene_expression <- expr['Trem2',] %>% 
  #t() %>% 
  as.data.frame()
```

```r
colnames(gene_expression) <- paste0(gene_name)
identical(colnames(wt3d),row.names(gene_expression))
```

```r
wt3d <- AddMetaData(
       object = wt3d,
       metadata = gene_expression
   )
```

```r
gene = wt3d@meta.data$Trem2
```

```r
wt3d@meta.data$trem2.group=ifelse( gene> median(gene),'high','low') 
```

```r
table(wt3d$trem2.group)
```

```r
Idents(wt3d) = 'trem2.group'
```

```r
diffgene  = FindMarkers(object = wt3d, ident.1 = 'high', min.pct = 0.25)
```

```r
head(diffgene,1)
```

```r
genelist <- as.numeric(diffgene[,2]) 
names(genelist) <- row.names(diffgene)
```

```r
OrgDb = 'org.Mm.eg.db'
```

```r
genelist_entrez <- genelist
names(genelist_entrez) <- bitr(names(genelist),
                               fromType="SYMBOL",toType="ENTREZID", 
                               OrgDb=OrgDb)[,2]  
genelist_entrez <- genelist_entrez[!is.na(names(genelist_entrez))]
```

```r
library(clusterProfiler)
library(org.Mm.eg.db)
```

```r
library(enrichplot)
library(clusterProfiler)
library(org.Mm.eg.db)
library(DOSE)
```

```r
geneList = genelist_entrez
```

```r
gene <- names(geneList)
```

```r
res2 <- enrichGO(gene, OrgDb = org.Mm.eg.db, ont = "BP")
res2 <- setReadable(res2, OrgDb = org.Mm.eg.db, keyType="ENTREZID")
```

```r
ego <- enrichKEGG(
 gene          = gene,
 keyType     = "kegg",
 organism   = 'mmu',
 pvalueCutoff      = 0.05,
 pAdjustMethod     = "BH",
 qvalueCutoff  = 0.05
)
```

```r
ego2 <- setReadable(ego, OrgDb = org.Mm.eg.db, keyType="ENTREZID")
```

```r
grep('Spp1',ego2@result$geneID)
```

```r
res = ego2@result[c(38,44,65,82,89,100,136),]
```

```r
write.csv(res,file = 'trem2.kegg.res.csv')
```

```r
ego2@result = res
```
