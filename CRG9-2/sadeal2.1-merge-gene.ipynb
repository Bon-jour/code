```r
library(dplyr)
```

```r
male.up = male  %>% filter(change == 'Up') %>%  rownames()
female.up = female  %>% filter(change == 'Up')%>%  rownames()
```

```r
intersect(male.up,female.up)
```

```r
male.d = male  %>% filter(change == 'Down') %>%  rownames()
female.d = female  %>% filter(change == 'Down')%>%  rownames()
```

```r
intersect(male.d,female.d)
```

```r
gene = c(intersect(male.up,female.up),intersect(male.d,female.d))
```
