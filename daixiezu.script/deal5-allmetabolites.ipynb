```r
mmu_kegg_compound <- read.csv('/data/workdirV6//KEGG_COMPOUND_PATHWAY_mmu.csv')

disease2gene=mmu_kegg_compound[, c("Pathway", "COMPOUND")]

disease2name=mmu_kegg_compound[, c("Pathway", "PATHWAY_MAP")]

cnumber = read.csv('/data/workdirV6/cnumber.csv',check.names = F,row.names = 1)
```

```r
kind = read.csv('./allmetebolit.kind.csv')
```

```r
kind = na.omit(cnumber[kind[,2],])
library(clusterProfiler)
x = enricher(kind, TERM2GENE=disease2gene, TERM2NAME=disease2name)
x@result$p.adjust	= x@result$pvalue
library(ggplot2)
dotplot(x, font.size=15, showCategory=10, title="Enrichment KEGG Top10") + scale_size(rang=c(5.20))
```
