```r
library(readxl)
metabolites = read_excel('./LM2021-12430-东部战区总医院-小鼠肠道组织/LM2021-12430-东部战区总医院-小鼠肠道组织-LCMS-EMDB(项目报告)/6.差异代谢物/差异代谢物.xlsx',sheet = 4)
colnames(metabolites)
metabolites1 = metabolites[,c('Metabolites','kegg','VIP','P-value','log2(FC)')]
library(dplyr)
colnames(metabolites1)[4:5] = c('pvalue','log2fc')
se = metabolites1 %>% filter(VIP >1) %>% subset(., abs(log2fc) > 1 & pvalue < 0.05)
name = se$Metabolites
```

```r
diff_metabolite = dataMatrix[name,]
```

```r
head(diff_metabolite,2)
```

```r
mycounts_filter = readRDS('../workdirV6/4.cor/mycounts_filter.rds')
dim(mycounts_filter)
diffgene = read.table('../workdirV6//3.enrichment/DEGsid/ALL_KO-CLPvsWT-CLP.txt',header =T,sep = ' ')
diffgene1 = subset(diffgene,abs(log2FoldChange) >0 & pvalue < 0.05)
funcgene = mycounts_filter[rownames(diffgene1),]
head(funcgene,2)
```

```r
dim(funcgene)
```

```r
library(igraph)
library(dplyr)
library(Hmisc)
```

```r
trt_id <- c('KO.CLP','KO.NC','WT.CLP','WT.NC') ## 定义样本的关键词，然后从样本名抓取处理的样本
split_metabolites <- list()
for (i in trt_id){
    split_metabolites[[i]] = dplyr::select(diff_metabolite,contains(i))
}
```

```r
split_funcgene = list()
for (i in trt_id){
    split_funcgene[[i]] = dplyr::select(funcgene,contains(i))
}
```

```r
merge_cor = list()
otu = list()
meta = list()
p = list()
r = list()

for (i in seq(1:4)){
    otu[[i]] = split_metabolites[[i]]
    meta[[i]] = split_funcgene[[i]]
    library(psych)
    merge_cor[[i]]<- corr.test(as.data.frame(t(otu[[i]])), as.data.frame(t(meta[[i]])), method = 'spearman')
    p[[i]] = merge_cor[[i]]$p
    r[[i]] = merge_cor[[i]]$r
    r[[i]][p[[i]]>0.05|abs(r[[i]])<0.2] = 0
    r[[i]][is.na(r[[i]])]=0
    p[[i]][is.na(p[[i]])]=0

    outdir = paste0(getwd(),'/6.metabolites-funcgene')
    if(dir.exists(outdir)){
      print("dir exists")
    }else{
      dir.create(outdir)
    }
    write.csv(r[[i]],file = paste0(outdir,'/',trt_id[i],'.cor.r.csv'))
    write.csv(p[[i]],file = paste0(outdir,'/',trt_id[i],'.cor.p.csv'))
}
```
