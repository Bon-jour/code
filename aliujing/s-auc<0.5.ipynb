```r
library(pROC)
```

```r
data(aSAH)
```

```r
res <- roc(aSAH$outcome,aSAH$s100b,ci=T,auc=T)
```

```r
res
```

```r
pre <- glm(outcome ~s100b,family=binomial(link = "logit"),data = aSAH)
summary(pre)
```

```r
length(aSAH$outcome)
```

```r
aSAH$outcome
```

```r
predict. <- predict.glm(pre,type='response',newdata=aSAH)
#按照预测值为1的概率，>0.5的返回1，其余返回0
predict =ifelse(predict.>0.5,'Poor','Good')
#数据中加入预测值一列
predict
```
