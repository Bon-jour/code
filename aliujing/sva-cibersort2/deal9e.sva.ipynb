```r
count = merge(data1,data2,all=FALSE,by = 'Gene_name')
```

```r
count = count %>%  tibble::column_to_rownames(.,'Gene_name')
```

```r
Group = c(rep('con',31),rep('CI',43),rep('con',19),rep('CI',16))
```

```r
matrix = c(rep('Matrix1',74),rep('Matrix2',35))
```

```r
anno = data.frame(sample = colnames(count),
          group = Group,
          Matrix = matrix )
```

```r
rownames(anno) = anno[,1]
anno = anno[,-1]
```

```r
anno = anno %>%  dplyr::filter(group == 'CI')
```

```r
count = count[,rownames(anno)]
```

```r
exp = count

cg=names(tail(sort(apply(exp,1,sd)),5000))
n=exp[cg,]

library(pheatmap)

forplot = na.omit(n)
#da=log2(data1+1)
dat=t(scale(t(forplot))) # 'scale'可以对log(fpkm+1)数值进行归一化
#处理数据
dat[dat>2]=2
dat[dat<(-2)]= -2
dim(dat)
dat = dat[complete.cases(dat), ]
```

```r
library(pheatmap)
```

```r
pheatmap(dat,
         cluster_rows = TRUE,
         cluster_cols = TRUE,
         show_rownames=FALSE,
         show_colnames = FALSE,
          scale="none",
          #cluster_cols = F,
          fontsize_row = 10,
          fontsize_col = 10,
         annotation_col = anno,
          #color = colorRampPalette(c("navy", "white", "firebrick3"))(100),
          color = colorRampPalette(c("#00BCB6", "white", "#FE4300"))(100)#换颜色
         # angle_col = 45 #修改横轴坐标名倾斜度
)
```

```r
library("FactoMineR")
library("factoextra")
pca.plot = function(dat,col){
  
  df.pca <- PCA(t(dat), graph = FALSE)
  fviz_pca_ind(df.pca,
               geom.ind = "point",
               col.ind = col ,
               addEllipses = TRUE,
               legend.title = "Groups"
  )
}
pca.plot(dat,factor(anno$Matrix))
```

```r
batch = anno$Matrix
combat_edata1 = ComBat(dat=count, batch=batch,par.prior=TRUE,  prior.plots=FALSE) 
```
