```r
LS = subset(scRNA,idents = 'HBP')
```

```r
library(Seurat)
library(dplyr)
```

```r
table(LS$seurat_clusters)
```

```r
Idents(scRNA) = 'seurat_clusters'
```

```r
sce = scRNA
```

```r
allCells=names(Idents(sce))
allType = levels(Idents(sce))

choose_Cells = unlist(lapply(allType, function(x){
  cgCells = allCells[Idents(sce)== x ]
  cg=sample(cgCells,20)
  cg
}))

cg_sce = sce[, allCells %in% choose_Cells]
```

```r
LS = cg_sce
```

```r
test = CreateObject_fromSeurat(Seurat.object=LS,slot="counts",
                               cell_type="celltype",
                               data_source="UMI",
                               scale.factor =10^6,
                               Org = "Homo sapiens") 
```

```r
mt <- TransCommuProfile(object = test,
                        pValueCor = 0.5,
                        CorValue = 0.1,
                        topTargetCor=1,
                        p.adjust = 0.5,
                        use.type="median",
                        probs = 0.5,
                        method="weighted",
                        IS_core = TRUE,
                        Org = 'Homo sapiens')
```

```r
n <- mt@data$expr_l_r_log2_scale
pathway.hyper.list <- lapply(colnames(n), function(i){
    print(i)
tmp <- getHyperPathway(data = n, object = mt, cella_cellb = i, Org="Homo sapiens")
return(tmp)
})
myPub.df <- getForBubble(pathway.hyper.list, cella_cellb=colnames(n))
plotBubble(myPub.df)
```
