```r
rm(list=ls())

library(bitops)
library(RCurl)
library(stringr)
library(XML)

genes <- read.table("test.txt",header = T,stringsAsFactors = F)
genes$NCBI_url <- paste("https://www.ncbi.nlm.nih.gov/gene/?term=",genes$ENTREZID,sep="")
head(genes)

getNodesTxt <- function(html_txt1,xpath_p){
  els1 = getNodeSet(html_txt1, xpath_p)
  # 获得Node的内容，并且去除空字符：
  els1_txt <- sapply(els1,xmlValue)[!(sapply(els1,xmlValue)=="")]
  # 去除\n：
  str_replace_all(els1_txt,"(\\n )+","")
}
dealNodeTxt <- function(NodeTxt){
  ifelse(is.character(NodeTxt)==T && length(NodeTxt)!=0 , NodeTxt , NA)
}
```

```r
c <- 1 #无响应或请求超时初始值
for(i in 1:nrow(genes)){
temp <- try(getURL(genes[i,"NCBI_url"]),silent=FALSE) #
if('try-error' %in% class(temp))
{doc <-NULL
cat('第',i,'个失败！\n')
c <- c+1
}else{
doc <- temp
cat('第',i,'个成功！\n')
html_txt1 = htmlParse(doc, asText = TRUE)

genes[i,"Locustag"] <- dealNodeTxt(getNodesTxt(html_txt1,'//*[@id="summaryDl"]/dd[preceding-sibling::dt[text()="Gene symbol" and position()=1 ] ]'))
cat("写入Locus tag\n")
}
if(c==20000){stop()} #若getURL
Sys.sleep(1) #睡眠1秒
}
```

```r
write.table(genes,'soybean_test_result.txt',quote = F,row.names = F)
```
